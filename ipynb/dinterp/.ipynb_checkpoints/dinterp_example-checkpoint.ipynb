```python
%matplotlib inline
import numpy as np
import numpy.random as random
import scipy.ndimage.measurements as measurements
import scipy.ndimage.morphology as morphology
import fitsio
import scipy.spatial.distance as distance
import matplotlib.cm as cm
```

```python
def dinterp(image=None, ivar=None, alpha=1., m=2):
    # Create the x- and -y pixel position images
    dims= image.shape
    ndim= len(dims)
    
    # Find regions of not-positive inverse variance
    ivar_zero= (ivar <= 0.)
    (labels_zero, nlabels)= measurements.label(ivar_zero)
    label_slices= measurements.find_objects(labels_zero)
    
    # Initialize image
    new_image= np.copy(image)
    
    # For each separate region ...
    for label_slice in label_slices:
        # Expand the region by 2 pixels
        label_image= np.zeros(dims)
        label_image[label_slice]= 1
        label_image= morphology.binary_dilation(label_image, iterations=m)
        sub_slice= measurements.find_objects(label_image)[0]
    
        # Create sub-images
        sub_image= np.copy(image[sub_slice])
        sub_ivar= np.copy(ivar[sub_slice])
        sub_new= np.copy(image[sub_slice])
        
        # Now find edges, as the places into which the 
        # dilation expanded
        sub_label= np.copy(label_image[sub_slice])
        sub_edge= (sub_label > 0) & (sub_ivar > 0)
        sub_iedge= np.nonzero(sub_edge)
        npix=len(sub_iedge[0]) 
        
        # Calculate pairwise distances
        sub_iedge_float= np.float32(sub_iedge).transpose()
        sub_distance= distance.squareform(distance.pdist(sub_iedge_float, metric='euclidean'))
        
        # That will give us the signal-signal matrix
        ss= np.exp(-0.25*(sub_distance/alpha)**2) 
        
        # The noise-noise matrix is diagonal
        nn= np.diag(1./(sub_ivar[sub_iedge]))
        
        # Now construct (S+N)^{-1}
        spn= ss+nn
        ispn=np.matrix(spn).I
        ee= np.matrix(np.ones(npix))
        clambda_denom= 1./((ee*ispn*ee.T).getA())[0][0]
        edge_image= np.copy(sub_image[sub_iedge])
            
        # For each zero inverse variance pixel, calculate the offset
        sub_izero= np.int32(np.nonzero(sub_ivar <= 0.)).transpose()
        sub_izero_float= np.float32(sub_izero)
        nzero= sub_izero.shape[0]
        for indx in np.arange(nzero):
            cdistance= distance.cdist(sub_izero_float[indx, :].reshape(1,ndim), sub_iedge_float, metric='euclidean')
            css=np.exp(-0.25*(cdistance/alpha)**2)
            sip=np.matrix(css)
            clambda_numer= ((ee*ispn*sip.T).getA())[0][0]
            clambda= (clambda_numer-1.)*clambda_denom
            dip= ispn*(sip-clambda*ee).T
            cval= np.inner(dip.getA().flatten(), edge_image)
            sub_new[tuple(sub_izero[indx,:])]= cval
        
        new_image[sub_slice]= sub_new
        
    return new_image
```

```python
def plot_interp_1d(nbad=1):
    nx=18
    alpha=1.2
    xx= np.arange(nx)
    xcens= [4.2, 8.9, 12.4]
    image= np.zeros(nx)
    for xcen in xcens:
        image= image+ np.exp(-0.5*(xx-xcen)**2/alpha**2)

    fig, axx = plt.subplots(nrows=7, ncols=2, sharex=True, sharey=True)
    iout= 1
    for ax in axx.flatten():
        ax.plot(xx, image)
        ivar= np.ones(nx)*10000.
        ivar[iout:iout+nbad]=0.
        new_image=dinterp(image=image, ivar=ivar, m=2, alpha=alpha)
        ax.plot(xx, new_image)
        izero= np.nonzero(ivar == 0.)[0]
        ax.plot(xx[izero], image[izero], '.', color='red')
        iout= iout+1
```

```python
plot_interp_1d(nbad=1)
```

```python
plot_interp_1d(nbad=2)
```

```python
nx=30
ny=30
image= np.zeros([nx, ny])
alpha= 1.2
xcens=[17.4, 5.3, 20.7, 4.7]
ycens=[15.1, 11.2, 6.5, 25.6]
yy= np.outer(np.arange(ny), np.ones(nx))
xx= np.outer(np.ones(ny), np.arange(nx))
for (xcen, ycen) in zip(xcens, ycens):
    rr= np.sqrt((xx-xcen)**2+(yy-ycen)**2)
    image= image+ np.exp(-0.5*rr**2/alpha**2)

ivar= np.ones([nx, ny])*10000.
ivar[:, 18:20]= 0.

new_image= dinterp(image=image, ivar=ivar, alpha=alpha, m=2)

greys= cm.get_cmap(name='Greys')
fig, (ax1, ax2)= plt.subplots(nrows=1, ncols=2, sharey=True)
ax1.imshow(image, interpolation='nearest', cmap=greys, 
           shape= image.shape)
ax2.imshow(new_image, interpolation='nearest', cmap=greys, 
           shape= new_image.shape)
```
